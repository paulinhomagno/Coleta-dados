```python
import pandas as pd
import os
    
def arquivos_em_df (diretorio):
        
    list_path = os.listdir(diretorio)
    
    df0 = pd.DataFrame()
    
    for arquivo in list_path:
        
        caminho = os.path.join(diretorio, arquivo)
        df = pd.read_csv(caminho, sep = '|', encoding = 'latin-1')
        df_final = pd.concat([df0, df])
        df0 = df_final
        
    return df_final
```

```python
df_pessoas = df[['cod_pessoa','genero', 'data_nascimento']].drop_duplicates()
df_hobbies = list(df['hobbies'].unique())
df_clima = list(df['clima'].unique())
df_bebidas = list(df['bebida_favorita'].unique())
df_animal = list(df['animal_estimacao'].unique())
```

```python
for i in df_pessoas.itertuples():
    
    id_pessoa = i.cod_pessoa
    genero = i.genero
    dt = str(i.data_nascimento)
    
    try:
        query = f""" INSERT INTO tb_pessoa (id_pessoa, genero, data_nasc)
                    VALUES ('{id_pessoa}', '{genero}', '{dt}')"""
        
        connect.execute(query)
        
    except Exception as e:
        print(f'Não foi inserido o registro de {id_pessoa}. Erro: {e}')
```
